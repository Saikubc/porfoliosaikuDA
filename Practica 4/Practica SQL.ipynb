### Pregunta 14

```python
sSQL='''
SELECT 
  'NORTE' AS ORIENTACION,
  AVG(UNITPRICE) AS PRECIO_PROMEDIO_M2
FROM dataset
WHERE HASNORTHORIENTATION = 1

UNION

SELECT 
  'SUR' AS ORIENTACION,
  AVG(UNITPRICE) AS PRECIO_PROMEDIO_M2
FROM dataset
WHERE HASSOUTHORIENTATION = 1

UNION

SELECT 
  'ESTE' AS ORIENTACION,
  AVG(UNITPRICE) AS PRECIO_PROMEDIO_M2
FROM dataset
WHERE HASEASTORIENTATION = 1

UNION

SELECT 
  'OESTE' AS ORIENTACION,
  AVG(UNITPRICE) AS PRECIO_PROMEDIO_M2
FROM dataset
WHERE HASWESTORIENTATION = 1;


'''
pd.read_sql(sSQL, conn)
```

### Pregunta 1

```python
sSQL='''
SELECT d.ASSETID, d.PRICE, d.HASSWIMMINGPOOL, d.HASPARKINGSPACE, a.AGENCY_NAME
FROM dataset d
JOIN agencias a ON d.AGENCY_ID = a.AGENCY_ID
WHERE d.HASSWIMMINGPOOL = 1 AND d.HASPARKINGSPACE = 1;
''' 
pd.read_sql(sSQL, conn)
```

### Pregunta 2

```python
sSQL='''
SELECT d.ASSETID, a.AGENCY_NAME, d.PRICE, a.RATING
FROM dataset d
JOIN agencias a ON d.AGENCY_ID = a.AGENCY_ID
WHERE a.RATING > 4.5;
''' 
pd.read_sql(sSQL, conn)
```

### Pregunta 3

```python
sSQL='''
SELECT a.AGENCY_NAME
FROM agencias a
LEFT JOIN dataset d ON a.AGENCY_ID = d.AGENCY_ID
WHERE d.AGENCY_ID IS NULL
'''
pd.read_sql(sSQL, conn)
```

### Pregunta 4

```python
sSQL = '''
SELECT a.AGENCY_NAME, AVG(d.PRICE) AS avg_price
FROM agencias a
JOIN dataset d ON a.AGENCY_ID = d.AGENCY_ID
GROUP BY a.AGENCY_NAME
ORDER BY avg_price DESC
LIMIT 1
'''
pd.read_sql(sSQL, conn)
```

### Pregunta 5

```python
sSQL = '''
SELECT d.ASSETID, d.PRICE, d.HASSWIMMINGPOOL, d.HASPARKINGSPACE, a.AGENCY_NAME
FROM dataset d
LEFT JOIN agencias a ON d.AGENCY_ID = a.AGENCY_ID
'''
pd.read_sql(sSQL, conn)
```

### Pregunta 6

```python
sSQL = '''
SELECT a.AGENCY_NAME, COUNT(d.ASSETID) AS NUM_INMUEBLES
FROM agencias a
LEFT JOIN dataset d ON a.AGENCY_ID = d.AGENCY_ID
GROUP BY a.AGENCY_NAME
ORDER BY NUM_INMUEBLES DESC
'''
pd.read_sql(sSQL, conn)
```
